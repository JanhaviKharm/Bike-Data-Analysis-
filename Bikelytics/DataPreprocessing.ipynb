```python
station_ids = set(df_stations['legacy_id'])
station_ids
for i in station_ids:
    df_id_from = df[df['from_station_id'] == i]
    df_id_to = df[df['to_station_id'] == i]
    df_id = [df_id_from,df_id_to]
    df_id = pd.concat(df_id,axis=0, ignore_index=True)
    
    s = pd.to_datetime(df_id['start_time'])
    df_gb_id = s.groupby(s.dt.floor('d')).size().reset_index(name='count')
    #print(df_gb_id)
    df_gb_id = df_gb_id.rename(columns={'start_time': 'date', 'count': 'value'})
    df_gb_id['date'] = df_gb_id['date'].dt.strftime('%Y-%m-%d')
    dict_d = df_gb_id.set_index('date').to_dict()
    
    idx = pd.date_range('2021-01-01', '2021-12-31')
    s = pd.Series(dict_d.get('value'))
    s.index = pd.DatetimeIndex(s.index)
    s = s.reindex(idx, fill_value=0)
    df111 = pd.DataFrame(s).reset_index()
    
    df111 = df111.rename(columns={'index': 'date', 0: 'value'})
    df111['date'] = df111['date'].dt.strftime('%Y-%m-%d')
    df111.to_json("./stationwise_daily_ridership/daily_ridership_counts_"+str(i)+".json" ,orient='records',date_format="%Y-%m-%d")
```

```python
#incoming
for i in station_ids:
    df_id = df[df['to_station_id'] == i]
    df_res = df_id.groupby([df_id.start_time.dt.hour]).count()
    df_res = df_res['trip_id']
    df_res = df_res.reset_index()
    df_res = df_res.rename(columns={'start_time': 'hour', 'trip_id': 'trips'})
    
    dict_stat = df_res.set_index('hour').to_dict()
    idx = range(0,24)
    s = pd.Series(dict_stat.get('trips'))
    s = s.reindex(idx,fill_value=0)
    df_stat = pd.DataFrame(s).reset_index()
    df_stat = df_stat.rename(columns={'index': 'hour', 0: 'trips'})
    df_stat.to_json("./stationwise_incomingstats/daily_ridership_counts_"+str(i)+".json" ,orient='records')
```

```python
#outgoing
for i in station_ids:
    df_id = df[df['from_station_id'] == i]
    df_res = df_id.groupby([df_id.start_time.dt.hour]).count()
    df_res = df_res['trip_id']
    df_res = df_res.reset_index()
    df_res = df_res.rename(columns={'start_time': 'hour', 'trip_id': 'trips'})
    
    dict_stat = df_res.set_index('hour').to_dict()
    idx = range(0,24)
    s = pd.Series(dict_stat.get('trips'))
    s = s.reindex(idx,fill_value=0)
    df_stat = pd.DataFrame(s).reset_index()
    df_stat = df_stat.rename(columns={'index': 'hour', 0: 'trips'})
    df_stat.to_json("./stationwise_outgoingstats/daily_ridership_counts_"+str(i)+".json" ,orient='records')
```

```python
months = ['May']
for month in months:
    print(month)
    df_month = df
    
    df_month_male = df_month[df_month['gender'] == 'Male']
    df_month_male_gb = df_month_male.groupby(df_month_male['start_time'].dt.floor('d')).count()
    df_month_male_gb = df_month_male_gb.rename(columns={'trip_id': 'male'})
    #print(df_month_male_gb)
    
    df_month_female = df_month[df_month['gender'] == 'Female']
    df_month_female_gb = df_month_female.groupby(df_month_female['start_time'].dt.floor('d')).count()
    df_month_female_gb = df_month_female_gb.rename(columns={'trip_id': 'female'})
    #print(df_month_female_gb)
    
    df_month_other = df_month[df_month['gender'] == 'Other']
    df_month_other_gb = df_month_other.groupby(df_month_other['start_time'].dt.floor('d')).count()
    df_month_other_gb = df_month_other_gb.rename(columns={'trip_id': 'other'})
    #print(df_month_other_gb)
    
    df_month_sub = df_month[df_month['usertype'] == 'Subscriber']
    df_month_sub_gb = df_month_sub.groupby(df_month_sub['start_time'].dt.floor('d')).count()
    df_month_sub_gb = df_month_sub_gb.rename(columns={'trip_id': 'subscriber'})
    #print(df_month_sub_gb)
    
    df_month_cust = df_month[df_month['usertype'] == 'Customer']
    df_month_cust_gb = df_month_cust.groupby(df_month_cust['start_time'].dt.floor('d')).count()
    df_month_cust_gb = df_month_cust_gb.rename(columns={'trip_id': 'customer'})
    
    df_month_whole = pd.concat([df_month_male_gb, df_month_female_gb,df_month_other_gb,df_month_sub_gb,df_month_cust_gb], axis=1)
    
    break
```
